# 繪製圖形
## 1. 畫一個櫻花形狀的函數圖形

繪製的函數：$r = a * \sin(k \cdot \theta) + \cos(k \cdot \theta) + 0.5$

$\theta$ 範圍：$0 \to 2\pi$，精度為1000

```python
k = 8      # 決定花瓣數量
a = 0.7      # 控制花瓣的大小
b = 0.5    # 控制花瓣的內縮和外展

# 建立角度 theta 的範圍
theta = np.linspace(0, 2 * np.pi, 10000)

# 定義極座標方程式
r = a * np.sin(k * theta)**2 + b * np.cos(k * theta) + 0.5

# 將極座標轉換為直角座標
x = r * np.cos(theta)
y = r * np.sin(theta)

# 繪製花朵圖形
plt.figure(figsize=(8, 8))
plt.plot(x, y, color='pink', linewidth=2)

# 設定圖形標題和座標軸
plt.title(f"極座標花朵圖案 (k={k})", fontsize=20)
plt.xlabel("X 軸")
plt.ylabel("Y 軸")
plt.axis('equal') # 確保X和Y軸比例相同，使花朵不變形
plt.grid(True)
plt.show()
```
